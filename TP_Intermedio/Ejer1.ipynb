```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm
import numpy as np
```

```python
#1) Cargue los datos del archivo advertising.csv suministrado en un dataframe de Pandas.
datos = pd.read_csv("advertising.csv")
```

```python
#2) Realice un gráfico de dispersión de los valores de Sales vs los índices de los datos.
plt.scatter(datos.index, datos['Sales'], color='blue')
plt.title('Sales vs Índice de los Datos')
plt.xlabel('Índice')
plt.ylabel('Sales')
plt.show()
```

```python
# 3) Utilice el comando train_test_split para separar los datos en conjuntos de entrenamiento y prueba. La fracción de datos de prueba debe estar
# entre 20% y 30%. 
X = datos[['TV', 'Radio', 'Newspaper']]
y = datos['Sales']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=21)
```

```python
# 4) Determine un modelo de regresión lineal múltiple utilizando los datos de entrenamiento. Este modelo debe estimar Sales a partir de las 
# variables de entrada.
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
y_pred = linear_regressor.predict(X_test)
```

```python
# 5) Evalúe el desempeño del modelo de regresión lineal obtenido en el apartado anterior.
print("Error absoluto medio =", round(sm.mean_absolute_error(y_test, y_pred), 2)) 
print("Error cuadratico medio =", round(sm.mean_squared_error(y_test, y_pred), 2))
print("Puntuación R2 =", round(sm.r2_score(y_test, y_pred), 2))
print("Puntuación de varianza explicada =", round(sm.explained_variance_score(y_test, y_pred), 2)) 
```

```python
# Realice un gráfico de dispersión de datos pronosticados (eje vertical) vs. datos de prueba (eje horizontal) 
plt.scatter(y_pred, y_test, color='orange', label='Predicciones', alpha=0.5)
plt.scatter(y_test, y_pred, color='red', label='Predicciones', alpha=0.5)
```

```python
# Realice un gráfico de línea de los errores porcentuales del modelo sobre los datos de prueba.
error_porcentual = np.abs((y_test - y_pred) / y_test) * 100
plt.plot(range(len(y_test)), error_porcentual, marker='o', linestyle='-', color='blue')
plt.title('Gráfico de Errores Porcentuales')
plt.xlabel('Índice')
plt.ylabel('Error Porcentual (%)')
plt.show()
```
