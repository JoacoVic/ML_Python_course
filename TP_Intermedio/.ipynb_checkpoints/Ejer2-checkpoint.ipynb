```python
from sklearn.datasets import load_wine
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
# 1) Cargue los datos en un dataframe de Pandas.
wine = load_wine()
datos_originales = pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
# 2) Genere 2 datasets, uno con los datos originales y otro con los datos de entrada normalizados.
scaler = StandardScaler()
data_normalizada = scaler.fit_transform(wine.data)
datos_normalizados = pd.DataFrame(data_normalizada, columns=wine.feature_names)

#Utilice el comando train_test_split para separar ambos datasets en conjuntos de entrenamiento y prueba.
X = wine.data
y = wine.target
X_train_or, X_test_or, y_train_or, y_test_or = train_test_split(X, y, test_size=0.25, random_state=5)
X_train_nor, X_test_nor, y_train_nor, y_test_nor = train_test_split(datos_normalizados, y, test_size=0.25, random_state=5)
```

```python
# 3) Determine dos modelos de clasificación basados en regresión logística, uno con cada conjunto de entrenamiento. Evalúe el desempeño de cada modelo
# sobre el conjunto de prueba correspondiente. Para esto, determine la precisión (accuracy) y la matriz de confusión para cada modelo.

#Datos originales
rl_or = LogisticRegression()
rl_or.fit(X_train_or, y_train_or)
accuracy_or = rl_or.score(X_test_or, y_test_or)
print(f"Exactitud en los datos originales: {accuracy_or:.2f}")

y_pred_or = rl_or.predict(X_test_or)
matriz_confusion_or = confusion_matrix(y_test_or, y_pred_or)
plt.figure(figsize=(6, 4))
sns.heatmap(matriz_confusion_or, cmap='Blues')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión - Regresión Logística en Datos Originales')
plt.show()
print("Matriz de confusión:\n", matriz_confusion_or)
```

```python
#Datos normalizados
clasificador_nor = LogisticRegression()
clasificador_nor.fit(X_train_nor, y_train_nor)
accuracy_nor = clasificador_nor.score(X_test_nor, y_test_nor)
print(f"Exactitud en los datos originales: {accuracy_nor:.2f}")

y_pred_nor = clasificador_nor.predict(X_test_nor)
matriz_confusion_nor = confusion_matrix(y_test_nor, y_pred_nor)
plt.figure(figsize=(6, 4))
sns.heatmap(matriz_confusion_nor, cmap='Blues')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión - Regresión Logística en Datos Normalizados')
plt.show()
print("Matriz de confusión:\n", matriz_confusion_nor)
```

```python
# 4) Determine dos modelos de clasificación basados en KNN, uno con cada conjunto de entrenamiento. Utilice el mismo K y la misma opción de weights
# (‘uniform’ o ‘distance’) para ambos modelos. Evalúe el desempeño de cada modelo sobre el conjunto de prueba correspondiente. Para esto, determine
# la precisión (accuracy) y la matriz de confusión para cada modelo.

k = 5
# Datos originales
knn_or = KNeighborsClassifier(n_neighbors=k, weights='distance')
knn_or.fit(X_train_or, y_train_or)
accuracy_or_knn = knn_or.score(X_test_or, y_test_or)
print(f"Exactitud KNN en los datos originales: {accuracy_or_knn:.2f}")

y_pred_or_knn = knn_or.predict(X_test_or)
matriz_confusion_or_knn = confusion_matrix(y_test_or, y_pred_or_knn)
plt.figure(figsize=(6, 4))
sns.heatmap(matriz_confusion_or_knn, cmap='Blues')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión - KNN en Datos Originales')
plt.show()
print("Matriz de confusión:\n", matriz_confusion_or_knn)
```

```python
# Datos normalizados
knn_nor = KNeighborsClassifier(n_neighbors=k, weights='distance')
knn_nor.fit(X_train_nor, y_train_nor)
accuracy_nor_knn = knn_nor.score(X_test_nor, y_test_nor)
print(f"Exactitud KNN en los datos normalizados: {accuracy_nor_knn:.2f}")

y_pred_nor_knn = knn_nor.predict(X_test_nor)
matriz_confusion_nor_knn = confusion_matrix(y_test_nor, y_pred_nor_knn)
plt.figure(figsize=(6, 4))
sns.heatmap(matriz_confusion_nor_knn, cmap='Blues')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión - KNN en Datos Normalizados')
plt.show()
print("Matriz de confusión:\n", matriz_confusion_nor_knn)
```

```python
# 5) Determine dos modelos de clasificación basados en SVM, uno con cada conjunto de entrenamiento. Utilice kernel de tipo lineal para ambos modelos.
# Evalúe el desempeño de cada modelo sobre el conjunto de prueba correspondiente. Para esto, determine la precisión (accuracy) y la matriz de confusión
# para cada modelo.

#Dator originales
svm_or = SVC(kernel='linear')
svm_or.fit(X_train_or, y_train_or)
accuracy_or_svm = svm_or.score(X_test_or, y_test_or)
print(f"Exactitud SVM en los datos originales: {accuracy_or_svm:.2f}")

y_pred_or_svm = svm_or.predict(X_test_or)
matriz_confusion_or_svm = confusion_matrix(y_test_or, y_pred_or_svm)
plt.figure(figsize=(6, 4))
sns.heatmap(matriz_confusion_or_svm, cmap='Blues')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión - KNN en Datos Originales')
plt.show()
print("Matriz de confusión:\n", matriz_confusion_or_svm)
```

```python
# 2. SVM para el dataset normalizado
svm_nor = SVC(kernel='linear')
svm_nor.fit(X_train_nor, y_train_nor)

# Evaluar en el conjunto de prueba del dataset normalizado
accuracy_nor_svm = svm_nor.score(X_test_nor, y_test_nor)
print(f"Exactitud SVM en los datos normalizados: {accuracy_nor_svm:.2f}")

y_pred_nor_svm = svm_nor.predict(X_test_nor)
matriz_confusion_nor_svm = confusion_matrix(y_test_nor, y_pred_nor_svm)
plt.figure(figsize=(6, 4))
sns.heatmap(matriz_confusion_nor_svm, cmap='Blues')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión - KNN en Datos Normalizados')
plt.show()
print("Matriz de confusión:\n", matriz_confusion_nor_svm)
```
